```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
csv_lc = pd.read_csv('data/run1.csv', sep=';') #TI eval kit data, measures at 20 samples per sec
csv_lc['code_mean'] = csv_lc['Codes - AIN0'].rolling(window=6000).mean() #6000 samples = 5min

csv_temp = pd.read_csv('data/run1usbtemp.csv', error_bad_lines=False, encoding ='utf-8', sep=',') #env sensor every 5min


csv_temp = csv_temp.sample(replace=True, frac=6000).sort_values(by='i')
#truncate to length of loadcell test
csv_temp = csv_temp[:len(csv_lc)]

csv_temp['temp_mean'] = csv_temp['temp'].rolling(window=6000).mean() #6000 samples = 5min
csv_temp['i'] = np.arange(0, len(csv_lc))

y = csv_lc['code_mean']-np.mean(csv_lc['code_mean'])
plt.figure()
plt.plot(csv_temp['i'], y/np.max(y))

plt.plot(csv_temp['i'], csv_temp['temp_mean']-np.mean(csv_temp['temp_mean'])) 
```
